```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import random
import numpy as np
```

```python
np.set_printoptions(precision=6, suppress=True) # 소수점 여섯자리까지만 표현
```

```python
# 한국어가 학습이 되어져 있는 모델을 가지고 와서 써야한다.
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
```

```python
sentence = ['안녕하세요 반갑습니다.','임베딩이 잘될까요?']
embeding = model.encode(sentence)
embeding.shape # 768차원의 정보로 768개의 단어로 구성이 되어있다. (압축 기법이 들어가서 저렇게 되어져있다,)
```

```python
df = pd.read_csv('ChatbotData.csv')
df
```

```python
# 문자를 벡터화(인코딩) 시키는 작업 (임베)
import csv
with open('embeding.csv','w') as file:
    mapdata = []

    write = csv.writer(file)

    for temp in df['Q']:
        emb = model.encode(temp)
        mapdata.append(emb)

    write.writerows(mapdata)
```

```python
df1 = pd.read_csv('embeding.csv',header=None)
df1.head()
```

```python
# 유사도 구하기
def chatbot_text(text):
    em_result = model.encode(text)

    co_result = []
    for temp in range(len(df1)):
        data = df1.iloc[temp]
        co_result.append(cosine_similarity([data],[em_result])[0][0])

    df['cos'] = co_result
    df_result = df.sort_values('cos',ascending=False)
    r = random.randint(0,5)

    return df_result.iloc[r]

chatbot_text('넌 바보야')
```

```python
import time
start = time.time()
result = chatbot_text('넌 바보야')
print(result)
print('time : ', time.time()-start)
```
