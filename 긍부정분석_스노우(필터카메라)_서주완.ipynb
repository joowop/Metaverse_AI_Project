```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('app_v2.csv', index_col = 0)
df.head(10)
```

```python
df = df.iloc[3:]
df
```

```python
# 인덱스 새로 생성하기
df.reset_index(inplace = True)
df.head(10)
```

```python
df.drop(['index'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# 똑같은 댓글 (중복 데이터) 제거
df = df.drop_duplicates(subset=['text'])
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
# 단어 사전 만들기 : 파싱 (말뭉치를 만들어야한다.) text에 있는 문장들을 하나로 만들면 된다.
text_data = ''.join(list(df['text']))
text_data
```

```python
okt = Okt()
nouns = okt.nouns(text_data)
print(nouns)
```

```python
# 빈도수 체크
counter = Counter(nouns)
print(counter)
```

```python
noun_data = Counter({x:counter[x] for x in counter if len(x) > 1})
noun_data.most_common(10)
```

```python
# 벡터화
import re
from tqdm import tqdm

noun_result = [] # [ [], [], [] ]

for temp in tqdm(df['text']):
    han = re.compile('[^ㄱ-ㅣ가-힣]') # 정규표현식
    temp_result = han.sub(' ', temp) # 문장에서 한글만 남기고 제거
    temp_noun = okt.nouns(temp_result)
    result_noun = [x for x in temp_noun if len(x) > 1] # 각 댓글을 명사 단위로 쪼갠뒤 한글자 이상인것만 저장
    noun_result.append(result_noun)

print(noun_result)
```

```python
# 글자를 숫자로 바꿔버리기 (bag of word 벡터 생성 )
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(tokenizer = lambda x : x, lowercase = False)
bow_vect = count_vectorizer.fit_transform(noun_result)
word_list = count_vectorizer.get_feature_names_out()
print(word_list)
len(word_list)
```

```python
bow_vect.toarray()
```

```python
len(bow_vect.toarray()[0])
```

```python
# 단어별 전체 등장 횟수
count_list = bow_vect.toarray().sum(axis=0)
count_list
```

```python
word_count = dict(zip(word_list, count_list))
word_count
```

```python
# tf-idf 생성
from sklearn.feature_extraction.text import TfidfTransformer

tf_vectorizer = TfidfTransformer()
tf_idf = tf_vectorizer.fit_transform(bow_vect)
tf_idf.toarray() # 각각 단어의 가중치가 담겨져있는 tridf가 만들어진다. (학습 데이터 생성 완료)
```

```python
df.head()
```

```python
# score가 4점부터는 긍정(1)으로 4점미만은 부정(0)으로 label로 설정할것이다.
df['score_real'] = np.where(df['score'] >= 4, 1, 0)
df.head()
```

```python
df['score_real'].value_counts() # 부정적인게 더 많다
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split

x = tf_idf
y = df['score_real']

# 긍정과 부정 중에 부정이 많기 때문에 두가지의 비율을 맞춰주기 위해서 stratify를 쓴다.
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1, stratify = df['score_real'])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# 모델 학습 이진분류
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score

lr_model = LogisticRegression(random_state=1)
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)


print("Accuracy:", accuracy)
print("Precision:", precision)

```

```python
invert_index_vectorizer = {v:k for k, v in count_vectorizer.vocabulary_.items()}
invert_index_vectorizer
```

```python
# 긍정적인 리뷰를 만드는데 사용된 단어들 찾기 reverse=True , False로 바꾸면 부정적 단어를 볼수 있다.
from sqlalchemy import true
coef_pos_index = sorted(((value,index) for index,value in enumerate(lr_model.coef_[0])),reverse=True)
for coef in coef_pos_index[:20]:
    print(invert_index_vectorizer[coef[1]], coef[0])
```

```python
def comment_predict(text):
    sentence_test = text
    han = re.compile('[^ㄱ-ㅣ가-힣]')
    temp_result = han.sub(' ', sentence_test)
    temp_noun = okt.nouns(temp_result)
    result_noun = [x for x in temp_noun if len(x) > 1]
    vec_test = count_vectorizer.transform([result_noun])
    vec_test.toarray().sum()
    tf_test = tf_vectorizer.transform(vec_test)
    tf_test.toarray().sum()

    result = ''
    if lr_model.predict(tf_test)[0] == 1:
        result = '긍정'
    else:
        result = '부정'
    return result
```
