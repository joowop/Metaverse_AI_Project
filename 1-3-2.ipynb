```python
# *매개변수 이면 매개변수가 몇개가 들어와도 상관없다.
def a(*x):
    tmp = 0
    for i in x:
        tmp = tmp+i
    return tmp
b = a(3,4,5,6,7)
b
```

```python
# *없애고 array로 매개변수를 넣어주면 몇개를 넣어도 상관없다.
def a(x):
    tmp = 0
    for i in x:
        tmp = tmp+i
    return tmp
b = a([3,4,5,6,7])
b
```

### 함수 작업

```python
def w(x):
    import numpy as np
    colCnt = len(x[0])
    # rand() 안에 들어가는 것의 의미는 컬럼의 개수이다. 
    w = np.random.rand(colCnt)*7
    w = w.astype('int32')
    return w

def cal(data, w):
    total = np.dot(data, w)
    return total

# df.values 는 df를 array로 변경해준 값이다.
we = w(df.values)
cal(df.values, we)
```
